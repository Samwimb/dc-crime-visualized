```python
import pandas as pd
import numpy as py
import datetime as dt
```

```python
# defining the file path
filepath = "2018weatherdata_csv.csv"
filepath
```

```python
# Open the file
weather_df = pd.read_csv(filepath)
```

```python
weather_df.head()
```

```python
# Drop columns with no data
del weather_df['L']
del weather_df['M']
del weather_df['H']
del weather_df['MW']
del weather_df['MW_1']
del weather_df['MW_2']
del weather_df['MW_3']
del weather_df['AW']
del weather_df['AW_4']
del weather_df['AW_5']
del weather_df['AW_6']
del weather_df['W']
del weather_df['PCP01']
del weather_df['PCP06']
del weather_df['PCP24']
del weather_df['PCPXX']
del weather_df['SD']
```

```python
weather_df.head()
```

```python
# Drop the rows with no temp
weather_df['TEMP'].replace({'****': None},inplace =True)
weather_df['TEMP'] = pd.to_numeric(weather_df['TEMP'])
```

```python
weather_df['MIN'].replace({'***': None},inplace =True)
weather_df['MAX'].replace({'***': None},inplace =True)
weather_df['DATE'] = weather_df['DATE'].astype('datetime64[ns]')
weather_df
```

```python
weather_df.dropna(axis=0, how='any', thresh=None, subset=['TEMP'], inplace=True)
weather_df
```

```python
weather_mean = pd.DataFrame (weather_df.groupby(['DATE']).mean()['TEMP']).reset_index()
weather_max = pd.DataFrame (weather_df.groupby(['DATE']).max()['TEMP']).reset_index()
weather_min = pd.DataFrame (weather_df.groupby(['DATE']).min()['TEMP']).reset_index()
weather_summary = pd.merge(weather_mean, weather_max, on='DATE', how='inner')
weather_summary = pd.merge(weather_summary, weather_min, on='DATE', how='inner')
weather_summary
```

```python
weather_summary = weather_summary.rename(columns={'TEMP_x': 'AVG', 'TEMP_y': 'MAX', 'TEMP': 'MIN'})
weather_summary
```

```python
filepath='2018_lunarClean.csv'
lunar_df = pd.read_csv(filepath)
lunar_df = lunar_df.rename(columns={'Date': 'DATE'})
lunar_df['DATE'] = lunar_df['DATE'].astype('datetime64[ns]')
```

```python
weather_data = pd.merge(weather_summary, lunar_df, on='DATE', how='inner')
```

```python
weather_data.to_csv('2018_weather_complete.csv", index=True, header=True)
```
