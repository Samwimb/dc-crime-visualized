```python
# Dependencies
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Name of the CSV file
file = '2018_lunar.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
Jan_df = df[['Day','Jan']]
Feb_df = df[['Day','Feb']]
Mar_df = df[['Day','Mar']]
Apr_df = df[['Day','Apr']]
May_df = df[['Day','May']]
Jun_df = df[['Day','Jun']]
Jul_df = df[['Day','Jul']]
Aug_df = df[['Day','Aug']]
Sep_df = df[['Day','Sep']]
Oct_df = df[['Day','Oct']]
Nov_df = df[['Day','Nov']]
Dec_df = df[['Day','Dec']]

Jan_df['Date']= '2018-01-' + Jan_df['Day'].astype(str)
#Jan_df = Jan_df[np.isfinite(df['Jan'])]
Jan_df['Date']= pd.to_datetime(Jan_df['Date']) 
del Jan_df['Day']
Jan_df = Jan_df.rename(columns={'Jan': 'PctIllum'})
Jan_df.set_index('Date', inplace = True)

Feb_df['Date']= '2018-02-' + Feb_df['Day'].astype(str)
Feb_df = Feb_df[np.isfinite(df['Feb'])]
Feb_df['Date']= pd.to_datetime(Feb_df['Date']) 
del Feb_df['Day']
Feb_df = Feb_df.rename(columns={'Feb': 'PctIllum'})
Feb_df.set_index('Date', inplace = True)

Mar_df['Date']= '2018-03-' + Mar_df['Day'].astype(str)
Mar_df = Mar_df[np.isfinite(df['Mar'])]
Mar_df['Date']= pd.to_datetime(Mar_df['Date']) 
del Mar_df['Day']
Mar_df = Mar_df.rename(columns={'Mar': 'PctIllum'})
Mar_df.set_index('Date', inplace = True)

Apr_df['Date']= '2018-04-' + Apr_df['Day'].astype(str)
Apr_df = Apr_df[np.isfinite(df['Apr'])]
Apr_df['Date']= pd.to_datetime(Apr_df['Date']) 
del Apr_df['Day']
Apr_df = Apr_df.rename(columns={'Apr': 'PctIllum'})
Apr_df.set_index('Date', inplace = True)

May_df['Date']= '2018-05-' + May_df['Day'].astype(str)
May_df = May_df[np.isfinite(df['May'])]
May_df['Date']= pd.to_datetime(May_df['Date']) 
del May_df['Day']
May_df = May_df.rename(columns={'May': 'PctIllum'})
May_df.set_index('Date', inplace = True)

Jun_df['Date']= '2018-06-' + Jun_df['Day'].astype(str)
Jun_df = Jun_df[np.isfinite(df['Jun'])]
Jun_df['Date']= pd.to_datetime(Jun_df['Date']) 
del Jun_df['Day']
Jun_df = Jun_df.rename(columns={'Jun': 'PctIllum'})
Jun_df.set_index('Date', inplace = True)

Jul_df['Date']= '2018-07-' + Jul_df['Day'].astype(str)
Jul_df = Jul_df[np.isfinite(df['Jul'])]
Jul_df['Date']= pd.to_datetime(Jul_df['Date']) 
del Jul_df['Day']
Jul_df = Jul_df.rename(columns={'Jul': 'PctIllum'})
Jul_df.set_index('Date', inplace = True)

Aug_df['Date']= '2018-08-' + Aug_df['Day'].astype(str)
Aug_df = Aug_df[np.isfinite(df['Aug'])]
Aug_df['Date']= pd.to_datetime(Aug_df['Date']) 
del Aug_df['Day']
Aug_df = Aug_df.rename(columns={'Aug': 'PctIllum'})
Aug_df.set_index('Date', inplace = True)

Sep_df['Date']= '2018-09-' + Sep_df['Day'].astype(str)
Sep_df = Sep_df[np.isfinite(df['Sep'])]
Sep_df['Date']= pd.to_datetime(Sep_df['Date']) 
del Sep_df['Day']
Sep_df = Sep_df.rename(columns={'Sep': 'PctIllum'})
Sep_df.set_index('Date', inplace = True)

Oct_df['Date']= '2018-10-' + Oct_df['Day'].astype(str)
Oct_df = Oct_df[np.isfinite(df['Oct'])]
Oct_df['Date']= pd.to_datetime(Oct_df['Date']) 
del Oct_df['Day']
Oct_df = Oct_df.rename(columns={'Oct': 'PctIllum'})
Oct_df.set_index('Date', inplace = True)

Nov_df['Date']= '2018-11-' + Nov_df['Day'].astype(str)
Nov_df = Nov_df[np.isfinite(df['Nov'])]
Nov_df['Date']= pd.to_datetime(Nov_df['Date']) 
del Nov_df['Day']
Nov_df = Nov_df.rename(columns={'Nov': 'PctIllum'})                                
Nov_df.set_index('Date', inplace = True)

Dec_df['Date']= '2018-12-' + Dec_df['Day'].astype(str)
Dec_df = Dec_df[np.isfinite(df['Dec'])]
Dec_df['Date']= pd.to_datetime(Dec_df['Date']) 
del Dec_df['Day']
Dec_df = Dec_df.rename(columns={'Dec': 'PctIllum'})                                
Dec_df.set_index('Date', inplace = True)



```

```python
Illum_df = pd.concat([Jan_df, Feb_df, Mar_df, Apr_df, May_df, Jun_df, Jul_df, Aug_df, Sep_df, Oct_df, Nov_df, Dec_df])
Illum_df
```
