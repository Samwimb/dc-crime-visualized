```python
import pandas as pd
import numpy as py
import datetime as dt
```

```python
# defining the file path
filepath1 = "Crime_Incidents_in_2018.csv"
filepath2 = "2018_weather_complete.csv"
```

```python
# Open the file
crime_df = pd.read_csv(filepath1)
weather_df = pd.read_csv(filepath2)
```

```python
crime_df['DATE'] = pd.to_datetime(crime_df['REPORT_DAT']).dt.date
crime_df['DATE'] = crime_df['DATE'].astype('datetime64[ns]')
weather_df['DATE'] = weather_df['DATE'].astype('datetime64[ns]')
weather_df = weather_df.drop([weather_df.columns[0]] ,  axis='columns')
weather_df.set_index('DATE', inplace=True)
```

```python
crime_weather_df = pd.merge(crime_df, weather_df, on='DATE', how='left')
```
